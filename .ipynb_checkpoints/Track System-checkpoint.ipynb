```python
import pandas as pd

#Reading Customer_Tracks.csv
dataset = pd.read_csv('./Customer_Tracks.csv')

#Showing first 40 rows of the dataset
dataset.head(40)
```

```python
#Storing Song Names in 2D array, in which each array stores songs listened by an individual user
currIndex = 0
tracks = [[dataset.loc[0][1]]]
currInvoice = dataset.loc[0][0]
for i in range(1, len(dataset)):
    if(dataset.loc[i-1][0] != dataset.loc[i][0]):
        currIndex = currIndex + 1
        tracks.append([])
    tracks[currIndex].append(dataset.loc[i][1])

#Deleting Duplicate Songs in same array
for i in range(0, len(tracks)):
    tracks[i] = list(set(tracks[i]))

#Printing first 5 rows
print(tracks[0:5])
```

```python
###Normalizing the Songs Names (Deleting \xa0 and Movie Name)###
#Extracting Song Names Only i.e. 
#Kabootar Ja Ja Ja - Maine Pyar Kiya (1989) => Kabootar Ja Ja Ja
#Mouka Milenga To Hum - Dilwale (1993) => Mouka Milenga To Hum

for i in range(0, len(tracks)):
    for j in range(0, len(tracks[i])):
        songName = tracks[i][j]
        songName = songName.split("-")[0]
        tracks[i][j] = songName[0:len(songName)-1]

print(tracks[0:5])
```

```python
#Creating Two Dictionary
#1. Assigning Numbers to Song and Accessing Those Numbers Using Song Name
#2. 2nd Dictionary for accessing songs from those numbers

songNo = 1

SongsToNum = {}
NumToSongs = {}

for i in range(0, len(tracks)):
    for j in range(0, len(tracks[i])):
        songName = tracks[i][j]    
        if songName not in SongsToNum.keys():
            SongsToNum[songName] = songNo
            NumToSongs[songNo] = songName
            songNo = songNo + 1
print("Songs to Number Dictionary:")
print(SongsToNum)
print("\n\nNumbers to Songs Dictionary:")
print(NumToSongs)
```

```python
#Converting DataSet from Songs Name to Numbers so Comparison can be done faster and Model is created Faster
for i in range(0, len(tracks)):
    for j in range(0, len(tracks[i])):
        tracks[i][j] = SongsToNum[tracks[i][j]]

print(tracks[0:5])
```

```python
from apyori import apriori

rules = apriori(tracks, min_support = 0.0003, min_confidence = 0.1, min_lift = 3, min_length = 2, max_length=3)

results = list(rules)

pdResults = pd.DataFrame(results)

pdResults.head(5)
```

```python
#Extracting Confidence from the Rows as We need to Design Reccomendation System so Only Confidence is required to
#check the relation of an element with another element.

confidence_list = []

for i in range(0, len(results)):
    for j in range(0, len(results[i].ordered_statistics)):
        confidence_list.append(results[i].ordered_statistics[j])

pdConfidence = pd.DataFrame(confidence_list)
pdConfidence.head(5)
```

```python
#Bubble Sort for Arranging the array in descending order w.r.t confidence

def bubbleSort(array):
  for i in range(len(array)):
    for j in range(0, len(array) - i - 1):
      if array[j].confidence < array[j + 1].confidence:
        temp = array[j]
        array[j] = array[j+1]
        array[j+1] = temp
```

```python
#Arranging Descendingly so Relations with more confidence can be pick first
bubbleSort(confidence_list)

pdConfidence = pd.DataFrame(confidence_list)
pdConfidence.head(5)
```

```python
#Give Suggestion Suggest the Songs according to the confidence
#parameters:
#songList => Set of Songs to be listened.
#limit => How much relations to be displayed.

def giveSuggestion(songList, limit):
    
    songSet = set()
    for song in songList:
        songSet.add(SongsToNum[song])
    for i in range(0, len(confidence_list)):
        if(limit==0):
            break
        if(songSet == confidence_list[i].items_base):
            print("There is ", round(confidence_list[i].confidence*100,2), "% chance of user to listen: ")
            for num in confidence_list[i].items_add:
                print(NumToSongs[num])
            print('\n')
            limit = limit-1

giveSuggestion({"Aaja Re"},6)
```
